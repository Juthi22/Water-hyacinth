```python
import cv2
import numpy as np
from PIL import Image, ImageFilter
import math
from numba import njit,prange
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/kaggle/input/waterhyacinth/Original Images/Water Lettuce (Pistia stratiotes)/20230818_112648.jpg',0)
imgp =img
img1 = img
img2 = img
img3 = img
img4 = img


im = Image.open("/kaggle/input/waterhyacinth/Original Images/Water Lettuce (Pistia stratiotes)/20230818_112648.jpg").convert("L")
im1 = im
```

```python
#***********average filtering***************
kernel = np.ones((3,3),np.float32)/25
dst = cv2.filter2D(imgp,-1,kernel)
#***********GaussianBlur***************
blur = cv2.GaussianBlur(img1,(3,3),0)
#***********median***************
median = cv2.medianBlur(img2,3)
#cv2.imwrite("Compressed___.png",median)
#***********Bilateral Filtering***************
bl = cv2.bilateralFilter(img3,3,75,75)
#***********min Filtering***************
i = im.filter(ImageFilter.MinFilter(size = 3))
#***********Max Filtering***************
i2 = im1.filter(ImageFilter.MaxFilter(size = 3))
# Create a grid to display the images
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# Display the original and filtered images
axes[0, 0].imshow(dst)
axes[0, 0].set_title("Average Filter", fontweight='bold')
axes[0, 0].axis('off')

axes[0, 1].imshow(blur)
axes[0, 1].set_title("Gaussian Filter", fontweight='bold')
axes[0, 1].axis('off')

axes[0, 2].imshow(median)
axes[0, 2].set_title("Median Filter", fontweight='bold')
axes[0, 2].axis('off')

axes[1, 0].imshow(bl)
axes[1, 0].set_title("Bileral Filter", fontweight='bold')
axes[1, 0].axis('off')

axes[1, 1].imshow(i)
axes[1, 1].set_title("Min Filter", fontweight='bold')
axes[1, 1].axis('off')

axes[1, 2].imshow(i2)
axes[1, 2].set_title("Max Filter", fontweight='bold')
axes[1, 2].axis('off')

plt.show()
```

```python
Y1 = np.square(np.subtract(img, dst)).mean()
print("AvegFilter MSE:", Y1)
Y2 = np.square(np.subtract(img, blur)).mean()
print("GauFilter MSE:", Y2)
Y3 = np.square(np.subtract(img, median)).mean()
print("MediFilter MSE:", Y3)
Y4 = np.square(np.subtract(img, bl)).mean()
print("BiliFilter MSE:", Y4)
Y5 = np.square(np.subtract(im, i)).mean()
print("MinFilter MSE:", Y5)
Y7 = np.square(np.subtract(im1, i2)).mean()
print("MaxFilter MSE:", Y7)
print("**********************************")
P1 = 10 * math.log10((255*255) / Y1)
print("AvegFilter PSNR:", P1)
P2 = 10 * math.log10((255*255) / Y2)
print("GauFilter PSNR:", P2)
P3 = 10 * math.log10((255*255) / Y3)
print("MediFilter PSNR:", P3)
P4 = 10 * math.log10((255*255) / Y4)
print("BiliFilter PSNR:", P4)
P5 = 10 * math.log10((255*255) / Y5)
print("MinFilter PSNR:", P5)
P7 = 10 * math.log10((255*255) / Y7)
print("MaxFilter PSNR:", P7)
```
